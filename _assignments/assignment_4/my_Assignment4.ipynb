### **(a)** For-loops and Conditional Statements  

Using the lists of movie ratings and titles below, where each rating corresponds directly to the title in the same position, create a new list of movies that are considered top picks (**those with ratings greater than or equal to 8**):   

- Use a for loop to iterate through both lists below and store the titles and ratings of movies considered top picks in two new lists:  

  - `top_picks_titles`: The titles of movies considered top picks.  

  - `top_picks_ratings`: The ratings of top pick movies.  
- Print the results of both lists.

````python
# Run this cell first
movie_ratings = [6.8, 8.3, 8.5, 5.1, 8.0, 3.5, 7.8, 8.4, 7.8, 2.8, 8.8]
movie_titles = ["Barbie", "Oppenheimer", "Spider-Man: Across the Spider-Verse", 
                "Morbius", "Dune", "The Emoji Movie", 
                "Everything Everywhere All at Once", "Avengers: Endgame", 
                "The Batman", "Cats", "Pulp Fiction"]
````

````python
# Write your answer here
top_picks_titles = []
top_picks_ratings = []

cut_off = 8

index = 0

for rating in movie_ratings:
    title = movie_titles[index]
    if rating >= cut_off:
        top_picks_titles = top_picks_titles + [title]
        top_picks_ratings = top_picks_ratings +[rating]
    index = index +1

print(top_picks_titles)
print(top_picks_ratings)
````

### **(c)** For-loops + Plotting

1. Create an array of x-values (`vec_x`) with 100 evenly spaced points between 0 and 2π using `np.linspace()`.

2. Create three y-vectors:
   - `vec_y1` using `np.sin(vec_x)`
   - `vec_y2` using `np.cos(vec_x)`
   - `vec_y3` using `np.sin(vec_x)` + `np.cos(vec_x)`

3. Create a list called `list_vectors` containing all three vectors: [`vec_y1`, `vec_y2`, `vec_y3`]

4. Write a for-loop that iterates through `list_vectors` and plots each y-vector against `vec_x` on the same plot.

5. Add axis labels, a title, and a legend (see the command `plt.legend`)

**Note**: The final output should be a single 2D plot showing all three functions plotted against the same x-values. It should look something like this:

<img src="figures/trig_functions.png" alt="drawing" width="350"/>

````python
# Write your answer here
vec_x = np.linspace(0, 2*np.pi, 100)

vec_y1 = np.sin(vec_x)
vec_y2 = np.cos(vec_x)
vec_y3 = np.sin(vec_x) + np.cos(vec_x)

list_vectors = [vec_y1, vec_y2, vec_y3]

for vector in list_vectors:
    plt.plot(vec_x, vector)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Trig Functions")
    plt.legend(["sinx", "cos(x)", "sin(x)+cos(x)"])
plt.show()
````
